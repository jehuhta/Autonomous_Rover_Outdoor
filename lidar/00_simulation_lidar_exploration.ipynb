```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example 1 lidar reading -- No object detected.
example1 = [
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf'
]

# Example 2 lidar reading -- Object detected.
example2 = [
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    4.188899993896484, 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf', 'inf',
    'inf', 'inf', 'inf', 'inf', 'inf', 'inf'
]

# Prints all the possible angles.
angles = list(range(-90,90,3))


```

```python
df = pd.DataFrame({
    'angle': angles,
    'distance': example2
})

df['distance'] = pd.to_numeric(df['distance'], errors='coerce').fillna(np.inf)

# Find obstacles within e.g. 2 metres
obstacles = df[df['distance'] < 2.0]
```

```python
THRESHOLD = 5.0
df['is_free'] = df['distance'] > THRESHOLD
```

```python
# Step 2 & 3: Find contiguous free regions and pick largest
# Label contiguous groups
df['gap_id'] = (df['is_free'] != df['is_free'].shift()).cumsum()
df
```

```python
gaps = df[df['is_free']].groupby('gap_id').agg(
    size=('angle', 'count'),
    center_angle=('angle', 'mean')
).reset_index()
```

```python
gaps
```

```python
largest_gap = gaps.loc[gaps['size'].idxmax()]
largest_gap
```
