<h3>Python Example to Read LiDAR Data</h3>

```python
#Install Python library:
# Install Python LiDAR library
pip install rplidar
```

```python
#Python Script
# Import LiDAR library
from rplidar import RPLidar

# Define the serial port where LiDAR is connected
PORT_NAME = '/dev/ttyUSB0'

# Create LiDAR object
lidar = RPLidar(PORT_NAME)

# Print health and info
print(lidar.get_info())
print(lidar.get_health())

# Start scanning
for scan in lidar.iter_scans():

    # scan contains many distance measurements
    for (_, angle, distance) in scan:

        # Print angle and distance
        print(f"Angle: {angle:.2f}  Distance: {distance:.2f}")

    # Break after first scan for testing
    break

# Stop LiDAR scanning
lidar.stop()

# Disconnect LiDAR
lidar.disconnect()
```
